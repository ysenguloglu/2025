#### Soru 1:
+ Bir banka müşterilerine ait 3 milyon telefon numarası verisini bir text dosyasında veriyor ve bizden telefonların sabit hat mı, cep telefonu hattı mı olduğunu, sabit hat ise hangi şehre ait olduğunu bulmamızı istiyor. Görevin yerine getirilmesi için nasıl bir yol izlersiniz?

```python
import pandas as pd
pd.set_option('display.max_rows', None)
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
def clean_phone_number(data = "Musteri_No.txt", phone_column="TELEFON"):
    df = pd.read_csv(data)
    df[phone_column] = df[phone_column].str.replace(" ","")
    df[phone_column] = df[phone_column].str.lstrip("0|+90")
    df = df[df[phone_column].str.len()==10].reset_index(drop=True)
    df["ALAN_KODU"] = df["TELEFON"].str[:3]

    return df

df = clean_phone_number()

def alan_kodu_ekle(data=df, alan_kodlari_location="alan_kodlari.xlsx"):
    
    alan_kodlari_df = pd.read_excel(alan_kodlari_location)
    alan_kodlari_df["Kod"] = alan_kodlari_df["Kod"].astype(str)

    alan_kodu_list = []
    for i in data["ALAN_KODU"]:
        if i[:1] == "5":
            alan_kodu_list.append("kisisel_cep")
        elif i in alan_kodlari_df["Kod"].values:
            alan_kodu_list.append(alan_kodlari_df[alan_kodlari_df["Kod"] == i]["Şehir"].values[0])
        else:
            alan_kodu_list.append("bilinmeyen")

    return alan_kodu_list

df["ALAN_KODU_SEHIR_EKLENMIS"] = alan_kodu_ekle()
df=df[~(df["ALAN_KODU_SEHIR_EKLENMIS"] == "bilinmeyen")].reset_index(drop=True)
df
```
